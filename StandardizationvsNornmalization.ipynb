```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
stock_df = pd.read_csv('S&P500_Stock_Data.csv')
stock_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Stock_df = scaler.fit_transform(stock_df)
```

```python
sns.pairplot(stock_df)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(stock_df.corr(), annot=True)
```
